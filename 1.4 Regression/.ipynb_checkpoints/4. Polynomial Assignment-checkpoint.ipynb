# Polynomial Regression - Assignment

What if your data is actually more complex than a simple straight line? Surprisingly,
you can actually use a linear model to fit nonlinear data. A simple way to do this is to
add powers of each feature as new features, then train a linear model on this extended
set of features. This technique is called Polynomial Regression

```python
# Some useful Imports
```

```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline
np.random.seed(42)
```

## Exercise - Fitting a Sine Wave

```python
# Show a Sine curve
X = np.linspace(-np.pi, np.pi, 201)
y=np.sin(X)

plt.plot(X, y, "b.")
plt.xlabel("x", fontsize=18)
plt.ylabel("y", rotation=0, fontsize=18)
#plt.axis([-3, 3, 0, 10])

plt.show()
```

### Assignment - Try different degrees of fit

```python
# Code for plotting predicted sinewave in different degrees
```

```python
#Predict and Display
# Start with values of 2 and change it to higher figures and observe changes in plot and R2 of fit
n_degree = 2


#Give X_values 100 values between -pi  and pi 
nwaves=2
X_values=np.linspace(-np.pi*nwaves, np.pi*nwaves, 100*nwaves).reshape(100*nwaves, 1)
y_values = np.sin(X_values)

# Transform X with defined  degree
poly_features = PolynomialFeatures(degree=n_degree, include_bias=False)
X_test = poly_features.fit_transform(X_values.reshape(-1, 1))

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_test, y_values)

# Predict for the transformed values
predictions = lin_reg.predict(X_test)

#Plot 
plt.figure(figsize=(12,6))
plt.plot(X_values, y_values, "b.", label="Original")
plt.plot(X_values, predictions, "r-", linewidth=2, label="Predictions")

plt.xlabel("x", fontsize=18)
plt.ylabel("y", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.title('Prediction Degree '+str(n_degree))
plt.show()


from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_values, predictions))
print('MSE:', metrics.mean_squared_error(y_values, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_values, predictions)))
print('R-squared:', metrics.r2_score(y_values, predictions))
```
